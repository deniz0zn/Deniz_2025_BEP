```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset/csv/BPI_Challenge_2019.csv")

# Assuming your DataFrame is called df and the timestamp column is 'time:timestamp'
# Parse the 'time:timestamp' column as datetime
df['time:timestamp'] = pd.to_datetime(df['time:timestamp'])

# Group by year and month and count the number of events per month
df['year_month'] = df['time:timestamp'].dt.to_period('M')  

events_per_month = df.groupby('year_month').size()  # Count events per month
print(events_per_month)
# Plot the distribution of events per month using a bar chart
plt.figure(figsize=(10, 6))
events_per_month.plot(kind='bar')
plt.title('Distribution of Events per Month')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
